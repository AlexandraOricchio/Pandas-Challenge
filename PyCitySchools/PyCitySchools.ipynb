```python
import pandas as pd
import os
```

```python
schools_csv=os.path.join("..","..","Resources","schools_complete.csv")
students_csv=os.path.join("..","..","Resources","students_complete.csv")

schools_df=pd.read_csv(schools_csv)
students_df=pd.read_csv(students_csv)
```

```python
schools_df.head()
```

```python
students_df.head()
```

```python
merged_df=pd.merge(students_df,schools_df,on="school_name",how="left")
merged_df.head()
```

```python
total_schools=merged_df['school_name'].unique()
total_schools=len(total_schools)
total_schools
```

```python
total_students=merged_df['Student ID'].count()
total_students
```

```python
total_budget=merged_df['budget'].unique()
total_budget=total_budget.sum()
total_budget
```

```python
avg_math=merged_df['math_score'].mean()
avg_math
```

```python
avg_reading=merged_df['reading_score'].mean()
avg_reading
```

```python
pass_math_df=merged_df.loc[merged_df['math_score']>69, :]
pass_math_count=pass_math_df['math_score'].count()
pass_math_avg=(pass_math_count/total_students)*100
pass_math_avg
```

```python
pass_reading_df=merged_df.loc[merged_df['reading_score']>69, :]
pass_reading_count=pass_reading_df['reading_score'].count()
pass_reading_avg=(pass_reading_count/total_students)*100
pass_reading_avg
```

```python
overall_passing_rate=(pass_math_avg + pass_reading_avg)/2
overall_passing_rate
```

```python
district_summary=pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students], "Total Budget":[total_budget], "Average Math Score":[avg_math], "Average Reading Score":[avg_reading], "% Pass Math":[pass_math_avg], "% Pass Reading":[pass_reading_avg], "Overall Passing Rate":[overall_passing_rate]})
district_summary['Total Students']=district_summary['Total Students'].map("{:,}".format)
district_summary['Total Budget']=district_summary['Total Budget'].map("${:,.2f}".format)
district_summary
```

```python
ss_grouped=merged_df.groupby(['type', 'budget','school_name'])
ss_grouped_df=pd.DataFrame(ss_grouped['Student ID'].count())
ss_grouped_df
```

```python
ss_df=ss_grouped_df.reset_index()
ss_df
```

```python
ss_df.columns
```

```python
order_ss_df=ss_df[['school_name','type','Student ID','budget']]
order_ss_df
```

```python
per_student_bud=(order_ss_df['budget']/order_ss_df['Student ID'])
order_ss_df['Per Student Budget']=per_student_bud
order_ss_df
```

```python
ss_grouped_avg=merged_df.groupby('school_name')
ss_grouped_avgDF=ss_grouped_avg['reading_score','math_score'].mean()
ss_grouped_avgDF
```

```python
merge_ss_df=pd.merge(order_ss_df,ss_grouped_avgDF,on='school_name',how='left')
merge_ss_df
```

```python
count_math=pass_math_df.groupby(['school_name'])
count_math_df=pd.DataFrame(count_math['math_score'].count())
count_math_df
```

```python
count_reading=pass_reading_df.groupby(['school_name'])
count_reading_df=pd.DataFrame(count_reading['reading_score'].count())
count_reading_df
```

```python
pass_count=pd.merge(count_math_df,count_reading_df,on='school_name',how='left')
pass_count_df=pass_count.rename(columns={"math_score":"Pass Math Count","reading_score":"Pass Reading Count"})
pass_count_df
```

```python
student_count=merge_ss_df.loc[:,['school_name','Student ID']]
student_count
```

```python
count_df=pd.merge(pass_count_df,student_count,on='school_name',how='left')
count_df
```

```python
per_passmath=(count_df['Pass Math Count']/count_df['Student ID'])*100
count_df['% Passing Math']=per_passmath
per_passreading=(count_df['Pass Reading Count']/count_df['Student ID'])*100
count_df['% Passing Reading']=per_passreading
count_df
```

```python
per_pass_df=count_df.loc[:,['school_name','% Passing Math','% Passing Reading']]
per_pass_df
```

```python
merge2_ss_df=pd.merge(merge_ss_df,per_pass_df,on='school_name',how='left')
merge2_ss_df
```

```python
passing_rate=(merge2_ss_df['% Passing Math']+merge2_ss_df['% Passing Reading'])/2
merge2_ss_df['Overall Passing Rate']=passing_rate
merge2_ss_df
```

```python
merge3_ss_df=merge2_ss_df.rename(columns={'school_name':'School Name','type':'School Type','Student ID':'Total Students','budget':'Total School Budget','reading_score':'Average Reading Score','math_score':'Average Math Score'})
merge3_ss_df
```
